```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("Data/Renewable_Energy_Usage_Sampled.csv")
df.head()
```

```python
os.makedirs("Output/plots", exist_ok=True)
```

VISUALIZATION 1: TREND OVER YEARS(LINE CHART)


```python
df.rename(columns={
    "Renewable_Energy_Usage (%)": "Renewable_Energy_Usage"
}, inplace=True)
```

```python
print(df.columns)
```

```python
print(df.columns.tolist())

```

```python
plt.figure(figsize=(10,6))

sns.lineplot(
    data=df,
    x="Year",
    y="Monthly_Usage_kWh"
)

plt.title("Trend of Monthly Renewable Energy Usage Over Years", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Monthly Energy Usage (kWh)")

plt.tight_layout()
plt.show()
```

Visualization 2: Region-wise Average Monthly Energy Usage (Bar Chart)

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=df,
    x="Region",
    y="Monthly_Usage_kWh",
    estimator="mean"
)

plt.title("Average Monthly Energy Usage by Region", fontsize=14)
plt.xlabel("Region")
plt.ylabel("Average Monthly Energy Usage (kWh)")

plt.tight_layout()
plt.show()
```

Visualization 3: Energy Source-wise Distribution (Pie Chart)

```python
plt.figure(figsize=(7,7))

energy_counts = df["Energy_Source"].value_counts()

plt.pie(
    energy_counts,
    labels=energy_counts.index,
    autopct="%1.1f%%",
    startangle=140
)

plt.title("Energy Source-wise Distribution")
plt.tight_layout()
plt.show()
```

Visualization 4: Urban vs Rural Average Energy Usage (Bar Chart)

```python
plt.figure(figsize=(7,5))

sns.barplot(
    data=df,
    x="Urban_Rural",
    y="Monthly_Usage_kWh",
    estimator="mean"
)

plt.title("Average Monthly Energy Usage: Urban vs Rural", fontsize=14)
plt.xlabel("Area Type")
plt.ylabel("Average Monthly Usage (kWh)")

plt.tight_layout()
plt.show()
```

Visualization 5: Income Level vs Monthly Energy Usage (Box Plot)

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x="Income_Level",
    y="Monthly_Usage_kWh"
)

plt.title("Monthly Energy Usage by Income Level", fontsize=14)
plt.xlabel("Income Level")
plt.ylabel("Monthly Energy Usage (kWh)")

plt.tight_layout()
plt.show()
```
